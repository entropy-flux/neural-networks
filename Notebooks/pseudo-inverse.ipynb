```python
from torch import Tensor
from torch import inverse

def moore_penrose_pseudo_inverse(X: Tensor) -> Tensor:
    return inverse(X.T @ X) @ X.T

def moore_penrose_linear_regression(features: Tensor, targets: Tensor) -> Tensor:
    return moore_penrose_pseudo_inverse(features) @ targets
```

```python
from torch import randn

N = 100
d = 5

X = randn(N, d) 
w_true = randn(d)
z = X @ w_true + 0.1 * randn(N, 1)  

w = moore_penrose_linear_regression(X, z)

print("True w:", w_true.squeeze())
print("Estimated w:", w.squeeze())
```
